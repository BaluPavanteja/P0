```python
import pandas as pd
from sqlalchemy import create_engine

# Creating DataFrames

transaction_logs_df = pd.DataFrame(transaction_data)
customers_df = pd.DataFrame(customer_data)

# Normalization
# 1. Create separate tables for products and orders
orders_df = transaction_logs_df[['order_id', 'customer_id', 'datetime', 'payment_type', 'ecommerce_website_name', 'payment_txn_id', 'payment_txn_success', 'failure_reason']].drop_duplicates().reset_index(drop=True)
products_df = transaction_logs_df[['product_id', 'product_name', 'product_category']].drop_duplicates().reset_index(drop=True)
order_items_df = transaction_logs_df[['order_id', 'product_id', 'qty', 'price']].reset_index(drop=True)

# MySQL connection parameters
host="localhost"
user="root"
password="Pavan@mysql29"
database="revstore"

# Create SQLAlchemy engine
engine = create_engine(f'mysql+mysqlconnector://{user}:{password}@{host}/{database}')

# Store DataFrames into MySQL tables
orders_df.to_sql('orders', con=engine, if_exists='replace', index=False)
products_df.to_sql('products', con=engine, if_exists='replace', index=False)
order_items_df.to_sql('order_items', con=engine, if_exists='replace', index=False)
customers_df.to_sql('customers', con=engine, if_exists='replace', index=False)

print("DataFrames have been successfully stored into MySQL tables.")
```
